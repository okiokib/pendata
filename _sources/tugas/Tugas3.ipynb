**Langkah 1: Mengimport Perpustakaan**

langkah pertama mengimpor pustaka yaitu NumPy, Pandas, dan Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Langkah 2: Mengimpor kumpulan data**

Pada tahap ini, kita akan melakukan proses impor dataset Bunga Iris yang tersimpan di GithubGist dengan nama file IrisDataset.csv. Kemudian hasil import dataset akan disimpan dalam variabel dataset. Setelah itu, kita akan menetapkan 4 variabel independen dari dataset ke variabel X dan variabel dependen 'species' ke variabel Y. Selanjutnya, kita akan menampilkan 5 baris pertama dari dataset tersebut menggunakan fungsi head(). Dengan cara ini, kita akan dapat melihat data awal dari dataset yang telah diimpor.

```python
dataset = pd.read_csv('https://gist.githubusercontent.com/okiokib/f40e40e4f076d1dbfcae8775be0c118e/raw/71572b2787bf31f6cdb65905afd8832a74db4c0c/irisDataset.csv')
X = dataset.iloc[:,:4].values
y = dataset['species'].values
dataset.head(5)
```

**Langkah 3: Memisahkan dataset menjadi set Pelatihan dan set Uji**

Setelah mendapatkan dataset Bunga Iris, langkah selanjutnya adalah membagi dataset menjadi kumpulan pelatihan dan pengujian. Dataset ini terdiri dari 150 baris data dengan masing-masing 50 baris dari 3 kelas yang diberikan secara berurutan. Oleh karena itu, agar pembagian data acak, kita perlu melakukan pemisahan dataset secara acak. Di sini, kita menggunakan parameter test_size=0.2 yang berarti 20% dari keseluruhan data akan digunakan sebagai data pengujian (test set), sedangkan 80% sisanya akan digunakan sebagai data pelatihan (train set) untuk melatih model klasifikasi Naive Bayes. Dengan membagi dataset menjadi data pelatihan dan pengujian, kita dapat menguji kinerja model yang dibuat pada data yang belum pernah dilihat sebelumnya dan menghindari terjadinya overfitting.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

**Langkah 4: Penskalaan Fitur**

Untuk meningkatkan kecepatan program, dataset dapat diperkecil ke dalam rentang yang lebih kecil dengan menggunakan teknik penskalaan fitur. Pada tahap ini, nilai dari variabel X_train dan X_test diperkecil ke dalam rentang nilai yang lebih kecil. Dengan melakukan penskalaan fitur, kita dapat memperkecil nilai-nilai dari setiap fitur pada dataset sehingga dapat mempercepat proses perhitungan pada model klasifikasi Naive Bayes. Dalam melakukan penskalaan fitur, kita dapat menggunakan beberapa metode seperti StandardScaler atau MinMaxScaler yang akan menormalisasi setiap fitur pada dataset. Dengan demikian, kita dapat mengoptimalkan kinerja model klasifikasi Naive Bayes pada dataset yang lebih kecil ini.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Langkah 5: Latih model Naive Bayes Classification pada Set Pelatihan**

Untuk melakukan klasifikasi pada dataset yang telah dibagi menjadi data pelatihan dan pengujian serta telah dilakukan penskalaan fitur, kita dapat menggunakan kelas GaussianNB dari pustaka sklearn.naive_bayes. Dalam melakukan klasifikasi, terdapat beberapa jenis model yang dapat digunakan seperti Bernoulli, Categorical, Multinomial dan Gaussian. Pada langkah ini, kita menggunakan model Gaussian dan menginisiasi objek kelas GaussianNB pada variabel classifier. Selanjutnya, kita menyesuaikan (fit) nilai X_train dan y_train pada objek kelas GaussianNB untuk tujuan pelatihan. Dengan melatih model klasifikasi pada dataset pelatihan, kita dapat menggunakan model tersebut untuk memprediksi nilai kelas pada dataset pengujian (test set).

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

**Langkah 6: Memprediksi hasil Test set**

Setelah model GaussianNB dilatih pada dataset pelatihan, kita dapat menggunakan metode classifier.predict() untuk melakukan prediksi nilai kelas pada dataset pengujian (test set). Hasil prediksi tersebut akan disimpan pada variabel y_pred. Dalam melakukan prediksi, model akan memproses nilai fitur dari setiap sampel pada test set dan menghasilkan nilai prediksi kelas berdasarkan model Gaussian yang telah dilatih pada dataset pelatihan. Dengan demikian, kita dapat menggunakan nilai prediksi pada variabel y_pred untuk mengevaluasi kinerja model klasifikasi pada dataset pengujian.

```python
y_pred = classifier.predict(X_test) 
y_pred
```

**Langkah 7: Matriks Kebingungan dan Akurasi**

Setelah melakukan prediksi pada dataset pengujian, kita dapat mengevaluasi kinerja model klasifikasi dengan menghitung akurasi model dan memplot confusion matrix. Confusion matrix adalah tabel yang menunjukkan jumlah prediksi yang benar dan salah pada suatu masalah klasifikasi ketika nilai sebenarnya dari test set diketahui. Tabel ini terdiri dari empat sel, yaitu True Positive (TP), False Positive (FP), True Negative (TN), dan False Negative (FN). True Positive (TP) merupakan jumlah prediksi yang benar dari kelas positif, sedangkan False Positive (FP) merupakan jumlah prediksi yang salah dari kelas negatif. True Negative (TN) merupakan jumlah prediksi yang benar dari kelas negatif, dan False Negative (FN) merupakan jumlah prediksi yang salah dari kelas positif.

Dalam hal ini, nilai akurasi dari model klasifikasi GaussianNB dihitung dan hasilnya ditampilkan. Selain itu, confusion matrix juga diproses dan ditampilkan dalam bentuk plot untuk memvisualisasikan kinerja model secara lebih jelas. Melalui plot confusion matrix, kita dapat melihat seberapa baik model dapat mengklasifikasikan data dengan benar pada setiap kelas.

**Hasil Akurasi :**

Hasilnya menunjukkan bahwa model memiliki akurasi sebesar 0.9333333333333333 atau sekitar 93%. Selain itu, matriks kebingungan juga diperbarui dan menunjukkan bahwa dari 30 sampel pada Test set, hanya satu sampel yang salah diklasifikasikan. Ini menunjukkan bahwa model yang dibangun cukup baik dalam mengklasifikasikan bunga berdasarkan 4 fitur karakteristik. Matriks kebingungan juga menunjukkan bahwa model memprediksi semua sampel dengan benar untuk 2 kelas (setosa dan versicolor), tetapi salah mengklasifikasikan 2 dari 10 sampel virginica. Meskipun demikian, hasil yang cukup baik telah dicapai oleh model klasifikasi Naive Bayes dalam memprediksi jenis bunga berdasarkan fitur-fiturnya.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred))
cm
```

**Langkah 8: Membandingkan Nilai Riil dengan Nilai Prediksi**

Pada tahap ini, kita membuat sebuah Pandas DataFrame untuk membandingkan nilai klasifikasi dari test set asli (y_test) dan hasil prediksi (y_pred). Meskipun tahap ini tidak terlalu informatif seperti confusion matrix dan biasanya digunakan dalam regresi untuk memeriksa keakuratan nilai prediksi, namun tahap ini dapat memberikan gambaran yang lebih jelas tentang kinerja model. Hasilnya menunjukkan bahwa terdapat satu prediksi yang salah, di mana bunga iris yang seharusnya diklasifikasikan sebagai virginica telah diprediksi sebagai versicolor. Meskipun demikian, secara keseluruhan, model klasifikasi Naive Bayes yang telah dilatih memiliki akurasi yang sangat baik, dengan akurasi mencapai 96,67%.

```python
df = pd.DataFrame({'Real Values':y_test, 'Predicted Values':y_pred})
df
```
