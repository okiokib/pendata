**A. Penggujian menggunakan metode Naive Bayes**

1. import library pandas, numpy, dan sklearn yang diperlukan

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

2. import dataset Heart Disease

```python
df = pd.read_csv('https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/master/dataset.csv')

data = df.shape[0]
df.head(data)
```

3. Mengecek nilai apakah terdapat nilai kosong pada data

```python
df_copy = df.copy()
df_copy.isnull().sum()
```

6. mengecek apakah semua type datanya sudah numerik/float

```python
df_copy.info()
```

7. Untuk melakukan normalisasi data dengan mengubah nilai-nilai ke dalam rentang antara 0 hingga 1, kita dapat menggunakan metode min-max. Dalam metode ini, nilai setiap data akan diubah dengan rumus (x-min) / (max-min). Terdapat beberapa metode normalisasi data lainnya seperti simple feature scaling atau z-score, namun perlu diingat bahwa kolom target tidak akan diikutsertakan dalam proses normalisasi (dalam kasus ini, kolom species).

```python
target_column = df_copy['target']
df_copy = df_copy[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']].copy()
df_copy = (df_copy - df_copy.min())/(df_copy.max() - df_copy.min())
df_copy['target'] = target_column.copy()
df_copy.head(data)
```

8. Pemotongan Data atau Data Slicing

> pisahkan data menjadi data training and data test. Kita dapat dengan mudah melakukan langkah ini menggunakan metode train_test_split() dari sklearn:

> Variabel features_train berisi data fitur untuk training, features_test berisi data fitur untuk test, target_train berisi target untuk training, dan target_test berisi target untuk test.

```python
features = df_copy.values[:,:13] 
target = df_copy.values[:,13]  
features_train, features_test, target_train, target_test = train_test_split(features,target, test_size = 0.33, random_state = 10) 
```

9. Implementasi Gaussian Naive Bayes

Setelah proses preprocessing data selesai dilakukan, langkah selanjutnya adalah menerapkan algoritma pembelajaran mesin atau machine learning. Pada kasus ini, kita akan menggunakan modul GaussianNB dari library sklearn. Dengan menggunakan modul ini, kita dapat membangun model yang mampu melakukan klasifikasi data dengan menggunakan pendekatan probabilitas dari distribusi Gaussian.

```python
clf = GaussianNB()
clf.fit(features_train, target_train)
target_pred = clf.predict(features_test)
```

10. Akurasi model Gaussian Naive Bayes

```python
accuracy = accuracy_score(target_test, target_pred, normalize = True)

print(f"Akurasi model Naive Bayes: {accuracy:.2%}")
```

## Kesimpulan

Berdasarkan nilai akurasi sebesar 79%, dapat disimpulkan bahwa data yang digunakan untuk melatih model telah memberikan hasil yang baik. Untuk menguji performa model yang telah dilatih, kita dapat membuat dataset uji secara acak dan melakukan pengujian untuk melihat seberapa baik kinerja model Gaussian Naive Bayes yang telah dibangun. **B. Pengujian menggunakan KNN**

1. import library pandas, numpy, dan sklearn yang diperlukan

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor 
from sklearn.preprocessing import LabelEncoder
```

2. import dataset Heart Disease

```python
df = pd.read_csv('https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/master/dataset.csv')

jumlah_baris = df.shape[0]
df.head(jumlah_baris)
```

3. Mengecek nilai apakah terdapat nilai kosong pada data

```python
df_copy = df.copy()
df_copy.isnull().sum()
```

4. Untuk melakukan normalisasi data dengan mengubah nilai-nilai ke dalam rentang antara 0 hingga 1, kita dapat menggunakan metode min-max. Dalam metode ini, nilai setiap data akan diubah dengan rumus (x-min) / (max-min). Terdapat beberapa metode normalisasi data lainnya seperti simple feature scaling atau z-score, namun perlu diingat bahwa kolom target tidak akan diikutsertakan dalam proses normalisasi (dalam kasus ini, kolom species).

```python
target_column = df_copy['target']
df_copy = df_copy[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']].copy()
df_copy = (df_copy - df_copy.min())/(df_copy.max() - df_copy.min())
df_copy['target'] = target_column.copy()
df_copy.head(jumlah_baris)
```

Langkah terakhir setelah melakukan inisialisasi objek knn dari kelas KNeighborsRegressor yang merupakan algoritma K-Nearest Neighbors untuk melakukan regresi, adalah melakukan randomisasi indeks dari dataset menggunakan np.random.permutation() dengan seed 1 untuk memastikan hasil yang konsisten. Selanjutnya, dataset dibagi menjadi 75% data untuk training dan 25% untuk testing dengan menggunakan variabel train_index. Data training dan data testing disimpan di variabel training_data dan testing_data, masing-masing.

Setelah model dilatih, prediksi dilakukan pada data testing dengan memanggil method predict() pada objek knn dengan argumen input berupa fitur-fitur dari data testing yang ingin diprediksi.

Setelah itu, akurasi model dihitung dengan mengukur berapa persen data testing yang berhasil diprediksi dengan menggunakan threshold default 0.5.

Terakhir, hasil akurasi model dicetak.

```python
knn = KNeighborsRegressor()

np.random.seed(1)
index_shuffled = np.random.permutation(df_copy.index)
shuffled_car_data = df_copy.loc[index_shuffled]

train_index = int(shuffled_car_data.shape[0] * 0.75)
training_data = shuffled_car_data.iloc[0:train_index]
testing_data = shuffled_car_data.iloc[train_index:]

knn.fit(training_data[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']], training_data['target'])

prediction = knn.predict(testing_data[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']])

accuracy = np.mean(np.abs(prediction - testing_data['target']) <= 0.5)

print(f"Akurasi model KNN: {accuracy:.2%}")
```

**Kesimpulan :**

Berdasarkan nilai akurasi sebesar 92%, dapat disimpulkan bahwa data yang digunakan untuk melatih model telah memberikan hasil yang baik. Untuk menguji performa model yang telah dilatih, kita dapat membuat dataset uji secara acak dan melakukan pengujian untuk melihat seberapa baik kinerja model KNN yang telah dibangun. Hal ini dapat membantu kita untuk mengetahui seberapa baik model dapat digunakan untuk memprediksi klasifikasi data baru secara akurat.
